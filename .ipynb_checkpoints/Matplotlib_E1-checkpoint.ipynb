```python
import matplotlib.pyplot as plt
%matplotlib inline

x=[1,2,3,4,5,6,7]
y=[25,45,67,11,65,80,55]

plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Days and Temperature')
plt.plot(x,y,color='blue',linewidth=3)
plt.plot(x,y, 'ro--')
plt.plot(x, y, color='blue',marker='o',  linestyle='dashed', linewidth=1, markersize=12,animated=True)






```
